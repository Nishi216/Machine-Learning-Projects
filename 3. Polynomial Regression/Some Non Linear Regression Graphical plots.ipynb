```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plotting the equation of y = 2x+3 - equation of degree 1

```python
x = np.arange(-5.0,5.0,0.1)
x
```

```python
y = 2*x + 3
y
```

```python
y_noise = 2 * np.random.normal(size=x.size)
y_noise
```

```python
y_data = y + y_noise
y_data
```

```python
plt.plot(x, y_data, 'yo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show();
```

### Plotting a graph for polynomial - $$ \ y = a x^3 + b x^2 + c x + d \ $$

```python
x = np.arange(-5.0,5.0,0.1)
y = 1*(x**3) + 1*(x**2) + 1*x + 3

y_noise = 15 * np.random.normal(size=x.size)
y_data = y + y_noise

plt.plot(x, y_data, 'yo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show();
```

### Quadratic function - $$ Y = X^2 $$


```python
x = np.arange(-5.0,5.0,0.1)
y = (x**2)

y_noise = 2 * np.random.normal(size=x.size)
y_data = y + y_noise

plt.plot(x, y_data, 'yo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show();
```

### Exponential function - $$ Y = a + b c^X$$

```python
x = np.arange(-5.0,5.0,0.1)
y = np.exp(x)

plt.plot(x, y, 'yo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show();
```

### Logarithm function - $$ y = \log(x)$$

```python
x = np.arange(5.0,10.0,0.1)
y = np.log(x)

plt.plot(x, y, 'yo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show();
```

### Sinusoidal or logistic function - $$ Y = a + \frac{b}{1+ c^{(X-d)}}$$

```python
x = np.arange(-5.0,5.0,0.1)
y = 1 - 4/(1+np.power(3, x-2))

plt.plot(x, y, 'yo')
plt.plot(x, y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show();
```
